```python
wine_array=wine.values
```

```python
wine_standard=scale(wine_array)
```

```python
wine_standard
```

```python
pca=PCA()
pca_values=pca.fit_transform(wine_standard)
```

```python
pca_values
```

```python
pca=PCA(n_components = 13)
pca_values=pca.fit_transform(wine_standard)
```

```python
pca_values
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals=3)*100)
var1
```

```python
pca.components_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(var1,color='blue')
```

```python
x=pca_values[:,0:1]
y=pca_values[:,1:2]
#z=pca_values[:,2:3]
plt.scatter(x,y)
```

```python
# K Mean clustering 
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
x=pca_values[:,0:3]  # 3 principal componant score to obtain 
x
```

```python
wcss=[]  # calculate within cluster sum of square distance

for i in range (1,11):  # range of no. of K (cluster)
    kmeans = KMeans (n_clusters=i, random_state=0)  
    kmeans.fit(x)   # x is normalize value for 3 PCA score 
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(10,6))
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.xlabel('Number of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
model=KMeans(n_clusters=3)
model.fit(x)
model.labels_
```

```python
md=pd.Series(model.labels_) # converting numpy array into pandas series object
db['Cluster']=md
db
```

```python
db.sort_values(by=['Cluster'])
```
